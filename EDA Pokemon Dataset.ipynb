```python
#printing some observations from the dataset in a table.
url = ''
pokemon = pd.read_csv('/content/drive/MyDrive/pokemon.csv') #including the pokemon dataset.
print('Dimensions of the pokemon dataset :',pokemon.shape)
pokemon.head(15)
```

# **Bar Graphs**

```python
#plotting a bar graph in same colored bars with "generation_id" in X-axis and "count" in Y-axis.
base_color = sb.color_palette()[9]
sb.countplot(data = pokemon, x = 'generation_id', color = base_color);
```

```python
#plotting a bar graph with different colored bars with "generation_id" in X-axis and "count" in Y-axis.
sb.countplot(data = pokemon, x = 'generation_id');
```

```python
base_color = sb.color_palette()[9]
gen_order = pokemon['generation_id'].value_counts()
print('Descending order of total number of pokemon in different "generation_id" :\n')
print(gen_order)  #printing total number of pokemon with different "generation_id" in descending order.

#plotting a graph of "generation_id" in descending order of total number of pokemon with different "generation_id".
gen_order = pokemon['generation_id'].value_counts().index
sb.countplot(data = pokemon, x = 'generation_id', color = base_color, order = gen_order);
```

```python
#plotting a graph with "type_1" in X-axis and "count" in Y-axis. ("type_1" names are densely shown)
base_color = sb.color_palette()[8]
sb.countplot(data = pokemon, x = 'type_1', color = base_color);
```

```python
#plotting a graph with "type_1" in X-axis and "count" in Y-axis.
base_color = sb.color_palette()[8]
sb.countplot(data = pokemon, x = 'type_1', color = base_color)
plt.xticks(rotation=90); #"type_1" names are sparsely shown by rotating them."
```

```python
#plotting a graph with "count" in X-axis and "type_1" in Y-axis.
base_color = sb.color_palette()[8]
sb.countplot(data = pokemon, y = 'type_1', color = base_color);
```

# **Counting total number of missing data**

```python
#counting and printing the total number of missing data in pokemon dataset using in-built function of pandas.
print('Missing data in pokemon dataset :\n')
pokemon.isna().sum()
```

# **Bar Graph for missing data**

```python
na_counts = pokemon.isna().sum()
base_color = sb.color_palette()[7]

#plotting a graph for the missing data in the pokemon dataset
sb.barplot(x=na_counts.index.values, y=na_counts, color = base_color)
plt.xticks(rotation=90);
```

# **Pie Chart**

```python
#plotting a pie chart with "generation_id" of pokemon dataset.
sorted_counts = pokemon['generation_id'].value_counts()
print('Descending order of total number of pokemon in different "generation_id" :\n')
print(sorted_counts)
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False);
plt.axis('square');
```

# **Donut Chart**

```python
#plotting a donut chart with "generation_id" of pokemon dataset.
sorted_counts = pokemon['generation_id'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, wedgeprops = {'width' : 0.3});
plt.axis('square');
```

# **Histograms using matplotlib**

```python
#plotting histograms with matplotlib.
#plotting a histogram for "speed" of pokemon dataset.
print('Minimum Speed of Pokemon :',pokemon['speed'].min())
print('Maximum Speed of Pokemon :',pokemon['speed'].max())
base_color = sb.color_palette()[0]
plt.hist(data = pokemon, x = 'speed', color = base_color);
```

```python
#plotting a histogram for "speed" of pokemon dataset with 15 bins.
base_color = sb.color_palette()[0]
plt.hist(data = pokemon, x = 'speed', bins = 15, color = base_color);
```

```python
#plotting a histogram for "speed" of pokemon dataset with each bin of size 5.
base_color = sb.color_palette()[0]
bins_edges = np.arange(0, pokemon['speed'].max()+1, 5)
print('Bins with each bin of size 5 :\n')
print(bins_edges)  #printing the bins as an 1-D array.
plt.hist(data = pokemon, x = 'speed', bins = bins_edges, color = base_color);
```

```python
plt.figure(figsize = [15, 5]) #increasing figure size for subplots.
base_color = sb.color_palette()[0]

#plotting histogram on left with too-large bin size.
plt.subplot(1, 2, 1) #rows = 1, columns = 2, subplots = 1
bin_edges = np.arange(0, pokemon['speed'].max()+1, +5)
plt.hist(data = pokemon, x = 'speed', bins = bin_edges, color = base_color);

#plotting histogram on right with too-small bin size.
plt.subplot(1, 2, 2) #rows = 1, columns = 2, subplots = 2
bin_edges = np.arange(0, pokemon['speed'].max()++1, 1)
plt.hist(data = pokemon, x = 'speed', bins = bin_edges, color = base_color);
```

# **Histograms using seaborn**

```python
#plotting a histogram using seaborn.
#with the default curve depicting kernal density estimate(KDE) of pokemon dataset.
base_color = sb.color_palette()[2]
sb.distplot(pokemon['speed'], color = base_color);
```

```python
#plotting histogram without the curve depicting kernal density estimate(KDE) of pokemon dataset.
base_color = sb.color_palette()[2]
sb.distplot(pokemon['speed'], kde=False, color = base_color);
```

```python
#plotting a histogram by increasing the transparency of the bins.
base_color = sb.color_palette()[2]
bin_edges = np.arange(0, pokemon['speed'].max()+1, 5)
sb.distplot(pokemon['speed'], bins = bin_edges, kde = False, hist_kws = {'alpha' : 1}, color = base_color);
```
